# Statistics

### Descriptive Analytics using Pandas

```python
## import the libraries

import pandas as pd
import numpy as np

```

```python
## Prepare a dataset df to apply our statistics
```

```python
# Create a Dictionary of series with 3 fields Name, age and rating

#Create a Dictionary of series
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack','Lee','David','Gasper','Betina','Andres']),
     'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
     'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])}

#Create a DataFrame
df = pd.DataFrame(d)
print(df)
```

Central Tendency - the tendency for the values of a random variable to cluster round its mean, mode, or median.


```python
# Mean
## Mean calculation
# Calculates the average and displays for its respective fiels which is numeric alone
print(df.mean())
```

```python
# Mean is the average of all numbers and then divide the sum by the total count of numbers.
# The population mean is indicated by the Greek symbol µ
# mean is calculated on a distribution from a sample it is indicated by the symbol x̅ 
```

```python
## Advantage : The mean can be used for both continuous and discrete numeric data.

## Limitations of the mean:

The mean cannot be calculated for categorical data, as the values cannot be summed.

As the mean includes every value in the distribution the mean is influenced by outliers and skewed distributions.
```

```python
## Median - The median is the middle value in distribution when the values are arranged in ascending or descending order.
```

```python
# median
## median calculation
# Calculates the average and displays for its respective fiels which is numeric alone
print(df.median())
```

```python
## Odd case
Looking at the retirement age distribution (which has 11 observations), the median is the middle value, which is 57 years: 

54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60 

## Even case
When the distribution has an even number of observations, the median value is the mean of the two middle values. 
In the following distribution, the two middle values are 56 and 57, therefore the median equals 56.5 years: 

52, 54, 54, 54, 55, 56, 57, 57, 58, 58, 60, 60
```

```python
Advantage of the median:

The median is less affected by outliers and skewed data than the mean, and is usually the preferred measure of central tendency when the distribution is not symmetrical. 


Limitation of the median:

The median cannot be identified for categorical nominal data, as it cannot be logically ordered.
```

```python
 ## Mode - The mode is the most commonly occurring value in a distribution.## Mode
```

```python
# Mode
## Mode calculation
# Calculates the average and displays for its respective fiels which is numeric alone
print(df)
print('\n')

print(df.mode())
```

```python
# column mode of the dataframe
df.mode(axis=0)
```

```python
# mode of the specific column
df.loc[:,"Age"].mode()
#df.loc[:,"Rating"].mode()
```

Advantage of the mode:

Mode has an advantage over the median and the mean as it can be found for both numerical and categorical (non-numerical) data.


Limitation :
    It is also possible for there to be more than one mode for the same distribution of data, 
    (bi-modal, or multi-modal). The presence of more than one mode can limit the ability of the mode in 
    describing the centre or typical value of the distribution because a single value to describe the 
    centre cannot be identified.

```python
## find the minimum value
print(df.min())
```

```python
## find the Maximum value
print(df.max())
```

```python
# Standard deviation


print(df.std())
```

```python
# Draw a bell curve
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

# perfect bell curve or a normal distribution curve should have  mean as 0 and std as 1
# curve is symmetric

mu = 0 # mean
variance = 1 #variance
sigma = math.sqrt(variance)

## bell curve to plot between - 3 sigma till + 3 sigma
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

#plot the curve
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

### Correlation

```python
import pandas as pd

## read a path of file from URL
path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

## pass the url and store the data in a variable ans specify the column names
mpg_data = pd.read_csv(path, delim_whitespace=True, header=None,
            names = ['mpg', 'cylinders', 'displacement','horsepower',
            'weight', 'acceleration', 'model_year', 'origin', 'name'],
            na_values='?')
mpg_data
```

```python
mpg_data.info()
## inference - horsepower had 6 missing values which will be dropped during the correlation calculation
#  horsepower      392
```

```python
mpg_data.describe()
```

```python
## gives the correlation matrix, this is applicable only for numerical data
mpg_data.corr()
```

```python
## correlation between mpg and weight
mpg_data['mpg'].corr(mpg_data['weight'])
## inference string negative correlation
```

```python
# pairwise correlation
mpg_data.drop(['model_year', 'origin'], axis=1).corr(method='spearman')
```

```python
 import matplotlib.pyplot as plt
```

```python
mpg_data.drop(['model_year', 'origin'], axis=1).corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# plot correlated values
plt.rcParams['figure.figsize'] = [16, 6]

fig, ax = plt.subplots(nrows=1, ncols=3)

ax=ax.flatten()

cols = ['weight', 'horsepower', 'acceleration']
colors=['#415952', '#f35134', '#243AB5', '#243AB5']
j=0

for i in ax:
    if j==0:
        i.set_ylabel('MPG')
    i.scatter(mpg_data[cols[j]], mpg_data['mpg'],  alpha=0.5, color=colors[j])
    i.set_xlabel(cols[j])
    i.set_title('Pearson: %s'%mpg_data.corr().loc[cols[j]]['mpg'].round(2)+' Spearman: %s'%mpg_data.corr(method='spearman').loc[cols[j]]['mpg'].round(2))
    j+=1

plt.show()
```

```python
## covariance
```

```python
Covariance is a measure of how changes in one variable are associated with changes in a second variable. Specifically, 
covariance measures the degree to which two variables are linearly associated

## eg. S&P with stock price
```

```python
## gives the covariance matrix, this is applicable only for numerical data
mpg_data.cov()
```
