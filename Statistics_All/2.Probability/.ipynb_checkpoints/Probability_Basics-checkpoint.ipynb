# Probability Basics

```python
Experiment   : Tossing a Coin
Trails       : How many times its tossed
Event        : outcome of interest 
Sample space : Possible events ( head or tail)

Probability  : chance of event occuring
    
Probability (P) : Favourable outcome / possible outcome
    
note : Probability output can only be between 0 and 1
```

```python
## Example - Flip a coin Once
## There are two possible outcomesâ€”heads or tails.

```

```python
## Probability

## Possibility of head to occur  P(h) = 1/2 = 0.5
## P(h) = (# of ways h can happen) / (Total number of outcomes)

## Possibility of tail to occur  P(t) = 1/2 = 0.5
## head and tail both has 50% chances
```

```python
#Simulation of Coin Toss
 
import random
#simulations typically make use of random numbers 
 
n = 10 #change this value to change the sample size
heads = 0
tails = 0
 
for i in range(n):
    #print(i) # i value is between 0 - 9
    flip = random.randint(1,2) # get a random number between 1 and 2
    if flip == 1: # head
        heads = heads + 1
    else:         # tail
        tails = tails + 1
 
print('Heads:', heads)
print('Tails:', tails)
```

```python
## Example 2 - Roll a 6 sided dice
```

```python
# Outcomes : 1,2,3,4,5,6
# trail - 1
# all the elements has equal probability to occur
# P(1) = 1/6 ,P(2) = 1/6,P(3) = 1/6 ,P(4) = 1/6 , P(5) = 1/6 ,P(6) = 1/6 
```

```python
# more than one event
P(1 or 3) = (1+1)/6 = 2/6
```

```python
import random
import operator

random.seed()

ROLLED = {i: 0 for i in range(1, 7)} # gets the range from 1 till 6 as the dice has 6 sides
ITERATIONS = int(input('How many times would you like to roll the dice? ')) # get the run time input

# create a function probability
def probability():
    print("Calculation of probability: ")
    for key, count in ROLLED.items():
        #print("count",count)
        #print("ITERATIONS",ITERATIONS)
        print("\t{}: {:.2f}".format(key, count*100./ITERATIONS*1.))

for _ in range(ITERATIONS):
    ROLLED[random.randint(1, 6)] += 1

## call the function probaility
probability()
```
